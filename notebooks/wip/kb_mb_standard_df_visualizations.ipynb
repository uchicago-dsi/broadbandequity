```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
std_df = pd.read_csv('/tmp/neighborhood-data/neighborhood_df.csv')
```

```python
std_df.head()
```

```python
labels = ['Seattle', 'Denver', 'Portland', 'Washington, D.C.',
         'Boston', 'Albuquerque', 'Houston', 'Indianapolis', 'Los Angeles',
         'Phoenix', 'San Francisco', 'Austin', 'Dallas', 'San Jose', 'San Diego']
colors = [labels.index(i) + 1 for i in std_df['City Name']]
```

```python
plt.scatter(std_df['Avg household income'], std_df['% Broadband Access'], 
            c=colors, label=labels, alpha=0.5)
plt.title("% Broadband Access by Household Income")
plt.xlabel("Average Household Income")
plt.ylabel("% Broadband Access")
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().legend(labels,loc='lower right')
```

```python
fig, ax = plt.subplots()
for i in range(len(labels)):
    city_vals = std_df.loc[std_df['City Name'] == labels[i]]
    x, y = city_vals['Avg household income'], city_vals['% Broadband Access']
    ax.scatter(x, y, label=labels[i],
               alpha=0.5, edgecolors='none')

plt.title("% Broadband Access by Household Income")
plt.xlabel("Average Household Income")
plt.ylabel("% Broadband Access")
# plt.gca().spines['top'].set_visible(False)
# plt.gca().spines['right'].set_visible(False)
ax.legend(bbox_to_anchor=(1.05, 1.0))
ax.grid(True)

plt.show()
```

```python
fig, ax = plt.subplots()
for i in range(len(labels)):
    city_vals = std_df.loc[std_df['City Name'] == labels[i]]
    x, y = city_vals['% >25 College Degree'], city_vals['% Broadband Access']
    ax.scatter(x, y, label=labels[i],
               alpha=0.5, edgecolors='none')

plt.title("% Broadband Access by College Degree")
plt.xlabel("% of People over 25 with a Bachelor's Degree")
plt.ylabel("% Broadband Access")
ax.legend(bbox_to_anchor=(1.05, 1.0))
ax.grid(True)

plt.show()
```

```python
fig, ax = plt.subplots()
for i in range(len(labels)):
    city_vals = std_df.loc[std_df['City Name'] == labels[i]]
    x, y = city_vals['% Black'], city_vals['% Broadband Access']
    ax.scatter(x, y, label=labels[i],
               alpha=0.5, edgecolors='none')

plt.title("% Broadband Access by Race")
plt.xlabel("% of People who are Black")
plt.ylabel("% Broadband Access")
ax.legend(bbox_to_anchor=(1.05, 1.0))
ax.grid(True)

plt.show()
```
