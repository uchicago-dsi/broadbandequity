### Creating "cleaned" race columns

Cleaned columns will include:

*  Hispanic (of any race)
*  Non-Hispanic White
*  Non-Hispanic Black
*  Non-Hispanic Asian
*  Non-Hispanic American Indian
*  Non-Hispanic Other

```python
acs_df_race_clean['Hispanic (of any race)'] = acs_df_race_clean['H/L']
acs_df_race_clean['Non-Hispanic White'] = acs_df_race_clean['Not H/L: White alone']
acs_df_race_clean['Non-Hispanic Black'] = acs_df_race_clean['Not H/L: Black alone']
acs_df_race_clean['Non-Hispanic Asian'] = acs_df_race_clean['Not H/L: Asian alone']
acs_df_race_clean['Non-Hispanic American Indian'] = acs_df_race_clean['Not H/L: American Indian/Native alone']
```

```python
acs_df_race_clean['Non-Hispanic Other'] = acs_df_race_clean['Not H/L: Native Hawaiian / PI alone'] + acs_df_race_clean['Not H/L: Other race alone'] + acs_df_race_clean['Not H/L: Two or more races']
```

```python
acs_df_race_final = acs_df_race_clean[['Hispanic (of any race)',
                                       'Non-Hispanic White',
                                       'Non-Hispanic Black',
                                       'Non-Hispanic Asian',
                                       'Non-Hispanic American Indian',
                                       'Non-Hispanic Other'
                                      ]]
```
