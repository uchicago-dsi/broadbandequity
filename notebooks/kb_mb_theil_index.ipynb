```python
def theil_index(city_name):
    city_pop = std_df_no_duplicates.loc[std_df_no_duplicates['City Name'] == city_name]['Population'].sum()
    pop_share = city_pop / std_df_no_duplicates["Population"].sum()
    
    city_avg_broadband = std_df_no_duplicates.loc[std_df_no_duplicates['City Name'] == city_name]["% Broadband Access"].mean()
    broadband_quotient = city_avg_broadband / std_df_no_duplicates["% Broadband Access"].mean()
    
    log_quotient = np.log(broadband_quotient)
    
    theil = pop_share * broadband_quotient * log_quotient
    
    return theil
```

```python
theil_index('seattle')
```

```python
theil_vals = []
for i in std_df_no_duplicates["City Name"].unique():
    print(i,theil_index(i))
    theil_vals.append(theil_index(i))
    
theil_df = pd.DataFrame({'City_Name' : std_df_no_duplicates["City Name"].unique(), "Theil Index" : theil_vals})
```

```python
theil_df = theil_df.sort_values(by=['Theil Index'])
```

```python
labels = std_df_no_duplicates['City Name'].unique()

fig, ax = plt.subplots()
for i in range(len(labels)):
    city_vals = std_df_no_duplicates.loc[std_df_no_duplicates['City Name'] == labels[i]]
    x, y = city_vals['% Hispanic'] + city_vals['% Black'], city_vals['% Broadband Access']
    ax.scatter(x, y, label=labels[i],
               alpha=0.5, edgecolors='none')

plt.title("% Broadband Access by Race")
plt.xlabel("% of People who are Black or Hispanic")
plt.ylabel("% Broadband Access")
lgd = ax.legend(bbox_to_anchor=(1.05, 1.0))
ax.grid(True)

fig.savefig("broadband_vs_race.png", bbox_extra_artists=(lgd,), bbox_inches='tight')
plt.show()
```
